```python
import pandas as pd
```

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
```

```python
movies.head()
```

```python
movies[movies.duration >= 200]
```

```python
movies[movies.duration >= 200 and movies.genre == 'Drama']
```

```python
movies[(movies.duration >= 200) & (movies.genre == 'Drama')]
```

```python
movies[(movies.duration >= 200) | (movies.genre == 'Drama')]
```

```python
(movies.duration >= 200) & (movies.genre == 'Drama')
```

```python
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Biography')]
```

```python
movies.genre.isin(['Action','Drama',"Biography"])
```

```python
movies[movies.genre.isin(['Action','Biography','Drama'])]
```

```python
boolean = []
for drama in movies.genre:
    if drama == 'Drama':
        boolean.append(True)
    else:
        boolean.append(False)
```
